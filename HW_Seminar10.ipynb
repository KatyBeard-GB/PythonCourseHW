**Задача 44:** В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего
из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без
get_dummies?


```python
import pandas as pd
import random

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

**1 Способ**

С использованием метода pandas get_dummies:

```python
pd.get_dummies(data['whoAmI'])
```

**2 Способ**

С использованием функции OneHotEncoder():

```python
df2 = data
```

```python
from sklearn. preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore')
```

```python
encoder_df = pd.DataFrame(encoder. fit_transform(df2[['whoAmI']]). toarray ())
encoder_df.columns = ['human', 'robot']
encoder_df
```

**3 Способ**

С использованием цикла for:

```python
lst_human, lst_robot = [], []
for i in lst:
  if i == 'human':
    lst_human.append(1)
    lst_robot.append(0)
  else: 
    lst_human.append(0)
    lst_robot.append(1)
df3 = pd.DataFrame({'human':lst_human, 'robot':lst_robot})
df3
```
